## Importação dos dados de empréstimos

Entendendo o padrão de nomeação dos arquivos. Nota-se que para cada ano tem-se dois arquivos, os nomes terminam com o ano seguido do numeral 1 ou 2. Provavelmente devem corresponder ao primeiro e segundo semestres do ano.

Definiremos loop for para carregar esses datasets sem ter de mencioná-los um a um.

```python
# Imports
import pandas as pd
import numpy as np
```

```python
list_year = np.arange(2010, 2021).tolist()
list_num = [1, 2]

# Link padrão de todos os arquivos armazenados no github - todos se iniciam com a url abaixo
pattern_link = 'https://raw.githubusercontent.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/main/Dia_1-Importando_dados/Datasets/dados_emprestimos/emprestimos-'
```

```python
%%time
## Importa todos os datasets e os armazena em variáveis como:
# df_20101, df_20102, df_20111, df_20112, ... 
for year in list_year:
    for num in list_num:
        try:
            df = pd.read_csv(pattern_link + str(year) + str(num) +'.csv')
            globals()['df_' + str(year) + str(num)] = df
        except:
            break
```

### Verificando duplicatas

```python
# Verifca duplicatas
print('Linhas duplicadas: ', lib_loans.duplicated().sum())

# Removendo duplicatas
lib_loans = lib_loans.drop_duplicates().reset_index(drop=True)
print('Linhas duplicadas após processo: ', lib_loans.duplicated().sum())
```
